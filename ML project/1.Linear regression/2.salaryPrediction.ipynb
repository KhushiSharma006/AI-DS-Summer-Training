```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle

df = pd.read_csv("experience_salary_40_records.csv")

X = df[["Experience"]]
y = df["Salary"]

model = LinearRegression()

model.fit(X, y)

predictedSalary = model.predict([[50]])
print("Predicted Salary:", predictedSalary[0])

pickle.dump(model, model("SalaryPredicted.pkl", "wb"))

```
