```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle

data = {
    "Area" : [1000, 1200, 1500, 1800, 2000],
    "price" : [25,30,40,50,55]
}

df = pd.DataFrame(data)

X = df[["Area"]]
y = df["price"]

model = LinearRegression()

model.fit(X, y)

predicted_price = model.predict([[2100]])
print("Predicted price for area 2100:", predicted_price[0])

pickle.dump(model, open("house_model.pkl", "wb"))

# pickle is a Python module used to serialize and deserialize Python objects.
# Serialization = Converting a Python object into a byte stream so it can be stored in a file.
# Deserialization = Loading the object back into memory.

print("Model Saved Successfully")
```
