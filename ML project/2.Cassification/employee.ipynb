```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Employee.csv")
```

```python

df.fillna({"Salary": df["Salary"].mean()}, inplace = True)
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
avg_salary = df.groupby("Education")["Salary"].mean()
print(avg_salary)
```

```python
plt.bar(
    avg_salary.index, avg_salary.values
)

plt.xlabel("Eduction")
plt.ylabel("Salary")
```

```python
avg_salary = df.groupby("City")["Gender"].value_counts().mean()
empCount = df["City"].value_counts()
plt.pie(
        empCount,
        labels=empCount.index
     )

plt.title("Employee per City")
plt.show()
```

```python

sns.countplot(data=df, x="City", hue="Gender")

plt.title("Gender Count in Each City")
plt.xlabel("City")
plt.ylabel("Number of Employees")
plt.show()
```
