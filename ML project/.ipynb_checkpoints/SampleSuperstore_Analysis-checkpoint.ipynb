# SampleSuperstore Analysis
Answers to the first 20 questions using `SampleSuperstore.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("SampleSuperstore.csv", encoding="latin1")
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.head()
```

## 1. Total Orders

```python
df["Order ID"].nunique()
```

## 2. Unique Customers

```python
df["Customer ID"].nunique()
```

## 3. State with Highest Sales

```python
df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(1)
```

## 4. City with Maximum Profit

```python
df.groupby("City")["Profit"].sum().sort_values(ascending=False).head(1)
```

## 5. Category with Highest Sales

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False).head(1)
```

## 6. Sub-Category with Highest Profit

```python
df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).head(1)
```

## 7. Most Used Ship Mode

```python
df["Ship Mode"].value_counts().head(1)
```

## 8. Customer with Highest Orders

```python
df.groupby("Customer Name")["Order ID"].count().sort_values(ascending=False).head(1)
```

## 9. Segment with Highest Sales

```python
df.groupby("Segment")["Sales"].sum().sort_values(ascending=False).head(1)
```

## 10. Total Sales and Total Profit

```python
pd.Series({"Total Sales":df["Sales"].sum(),"Total Profit":df["Profit"].sum()})
```

## 11. Region with Maximum Profit

```python
df.groupby("Region")["Profit"].sum().sort_values(ascending=False).head(1)
```

## 12. Average Sales per Order

```python
df.groupby("Order ID")["Sales"].sum().mean()
```

## 13. Product with Highest Sales

```python
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(1)
```

## 14. Product with Highest Profit

```python
df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(1)
```

## 15. State with Highest Loss

```python
df.groupby("State")["Profit"].sum().sort_values().head(1)
```

## 16. Sub-Category with Lowest Profit

```python
df.groupby("Sub-Category")["Profit"].sum().sort_values().head(1)
```

## 17. Monthly Sales Trend

```python
monthly=df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly.plot(figsize=(10,4),marker="o",title="Monthly Sales Trend")
plt.show()
monthly
```

## 18. Yearly Profit Trend

```python
yearly=df.groupby(df["Order Date"].dt.year)["Profit"].sum()
yearly.plot(figsize=(8,4),marker="o",title="Yearly Profit Trend")
plt.show()
yearly
```

## 19. Month with Highest Sales

```python
df.groupby(df["Order Date"].dt.month_name())["Sales"].sum().sort_values(ascending=False).head(1)
```

## 20. Average Discount by Category

```python
df.groupby("Category")["Discount"].mean()
```
