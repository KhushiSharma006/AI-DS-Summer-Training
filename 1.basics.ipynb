```python
#with recursion
def factorial(n):
    if n==0 or n==1:
        return 1
    elif n<0:
        return "enter positive number"
    else:
        n*factorial(n-1)
n=eval(input("enter number :"))
print(factorial(n))
```

```python
# without recursion
def calFactorial():
    n = eval(input("ENTER A NUMBER:" ))

    fact =1
    for i in range (1, n+1):
        fact = fact*i
    print(f"Factorial of {n}: {fact}")



calFactorial()
    
```
