```python
import pandas as pd

df = pd.DataFrame({
    "Order_ID": range(101, 121),
    "Product": [
        "Laptop","Mobile","Tablet","Laptop","Mobile",
        "Tablet","Laptop","Mobile","Tablet","Laptop",
        "Mobile","Tablet","Laptop","Mobile","Tablet",
        "Laptop","Mobile","Tablet","Laptop","Mobile"
    ],
    "Category": [
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics"
    ],
    "City": [
        "Delhi","Mumbai","Lucknow","Delhi","Lucknow",
        "Mumbai","Delhi","Lucknow","Mumbai","Delhi",
        "Lucknow","Mumbai","Delhi","Lucknow","Mumbai",
        "Delhi","Lucknow","Mumbai","Delhi","Lucknow"
    ],
    "Sales": [
        65000,30000,25000,70000,32000,
        27000,68000,35000,26000,72000,
        33000,28000,75000,34000,29000,
        76000,36000,30000,78000,37000
    ],
    "Profit": [
        12000,5000,4500,13000,5500,
        4700,12500,6000,4800,14000,
        6200,5000,14500,6300,5200,
        15000,6500,5400,15500,6800
    ],
    "Quantity": [
        5,10,8,6,9,
        7,5,11,8,6,
        10,7,5,9,8,
        4,12,9,5,10
    ]
})

print(df)
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    title="Sales vs Profit"
)

fig.show()
```

### reset_index() is a Pandas function that converts the index back into a normal column and assigns a new default integer index (0, 1, 2, ...).

```python
# 1. Create a bar chart showing the total Sales for each Product.
tot_Sales = df.groupby("Product")["Sales"].sum().reset_index()

fig = px.bar(
    tot_Sales,
    x = "Product",
    y = "Sales",
    title  = "Total sales for each product",
    color = "Product"
)
fig.show()
```

```python
# 2. Create a pie chart showing the Sales distribution by Product.

sales_product = df.groupby("Product")["Sales"].sum().reset_index()

px.pie(
    sales_product,
    names = "Product",
    values = "Sales",
    title = "distribution by Product"
)
```

```python
# 3. Create a scatter plot between Sales and Profit, using Quantity as the marker size and Product as the color
px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Product",
    size="Quantity",
    hover_data=["City"]
)
```

```python
# 4. Create a histogram of Sales with 10 bins.
px.histogram(
    df,
    x = "Sales",
    nbins = 10
)
```

```python
# 5. Create a box plot to compare the Sales distribution of each Product.
px.box(
    df,
    x = "Sales",
    y = "Product"
)
```

```python
# 6. Using groupby('City'), calculate the total Sales of each city and create a bar chart.

citySales = df.groupby("City")["Sales"].sum().reset_index()

px.bar(
    citySales,
    x = "City",
    y = "Sales",
    color = "City"
)
```

```python
# 7. Using groupby('Product'), calculate the average Profit of each product and create a line chart.
productProfit = df.groupby("Product")["Profit"].mean().reset_index()

px.line(
    productProfit,
    x = "Product",
    y = "Profit",
    # markers = True
)
```

```python
# 1. Create a bar chart showing the total Sales for each Product.
tot_Sales = df.groupby("Product")["Sales"].sum()

fig = px.bar(
    x = tot_Sales.index,
    y = tot_Sales.values,
    title  = "Total sales for each product",
    color = tot_Sales.index
)
fig.show()
```
