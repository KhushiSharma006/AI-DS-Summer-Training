```python
# Q47. Create a new column Profit = (Unit_Price - Unit_Cost) × Quantity.

df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

print(df.head())
```

```python
# Q48. Find the Top 10 most profitable transactions.

print(df.sort_values(by="Profit", ascending=False).head(10))
```

```python
# Q49. Analyze monthly sales trends using Sale_Date.

df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

monthly_sales = df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum()

print(monthly_sales)
```

```python
# Q50. Create an EDA Dashboard containing:
# • Total Sales
# • Total Profit
# • Region-wise Sales
# • Category-wise Sales
# • Monthly Sales Trend
# • Sales by Sales Representative
# • Customer Type Distribution
# • Payment Method Distribution
# • Sales Channel Distribution
# • Discount Analysis

df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

monthly_sales = df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum()

plt.figure(figsize=(18, 12))

plt.subplot(3,3,1)
df.groupby("Region")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Region-wise Sales")

plt.subplot(3,3,2)
df.groupby("Product_Category")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Category-wise Sales")

plt.subplot(3,3,3)
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")

plt.subplot(3,3,4)
df.groupby("Sales_Rep")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Sales by Sales Representative")

plt.subplot(3,3,5)
df["Customer_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Customer Type Distribution")
plt.ylabel("")

plt.subplot(3,3,6)
df["Payment_Method"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("")

plt.subplot(3,3,7)
df["Sales_Channel"].value_counts().plot(kind="bar")
plt.title("Sales Channel Distribution")

plt.subplot(3,3,8)
plt.hist(df["Discount"], bins=10)
plt.title("Discount Analysis")

plt.subplot(3,3,9)
plt.hist(df["Profit"], bins=10)
plt.title("Profit Distribution")

plt.tight_layout()
plt.show()

print("Total Sales:", df["Sales_Amount"].sum())
print("Total Profit:", df["Profit"].sum())
```
