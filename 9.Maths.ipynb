```python
import numpy as np

# Matrix A
A = np.array([
    [2, 4, 6],
    [8, 10, 12],
    [14, 16, 18]
])

# Matrix B
B = np.array([
    [1, 3, 5],
    [7, 9, 11],
    [13, 15, 17]
])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
```

```python
# 1. Find the shape, size and dimensions of both matrices.
print(np.shape(A))
print(np.shape(B))

print(np.size(A))
print(np.size(B))


print(A.ndim)
print(B.ndim)
```

```python
# 2. Perform element-wise addition of A and B.
A+B
```

```python
# 3. Perform element-wise subtraction (A - B).
A-B
```

```python
# 4. Perform element-wise multiplication.
A*B
```

```python
# 5. Perform matrix multiplication using np.matmul().
print(np.matmul(A, B))
```

```python
# 6. Find the transpose of both matrices.
print(A.T)
```

```python
# 7. Find the determinant of matrix A.
print(np.linalg.det(A))
```

```python
# 8. Find the inverse of matrix A (if possible).
if np.linalg.det(A) != 0:
    print(np.linalg.inv(A))
else:
    print("Inverse does not exist (Singular Matrix)")
```

```python
# 9. Find the trace (sum of diagonal elements) of both matrices.
print(np.trace(A))
print(np.trace(B))
```

```python
# 10. Find the maximum, minimum, mean and sum of matrix A.
print(np.max(A))
print(np.min(A))
print(np.mean(A))
print(np.sum(A))
```

```python
# 11. Find the row-wise sum and column-wise sum of matrix B.
print(np.sum(B, axis=1))   # Row-wise

print(np.sum(B, axis=0))   # Column-wise
```

```python
# 12. Extract:
# a) First row
print(A[0])

# b) Second column
print(A[:, 1])

# c) Last row
print(A[-1])

# d) Bottom-right element
print(A[-1, -1])
```

```python
# 13. Reshape matrix A into a (1,9) array and flatten matrix B.
print(A.reshape(1, 9))

print(B.flatten())
```

```python
# 14. Find the eigenvalues and eigenvectors of matrix A.
values, vectors = np.linalg.eig(A)

print(values)
print(vectors)
```
