```python
class Student:
    Courses = "Data Science"

    def __init__(self, rn, n, c, p, ds):   # Constructor
        self.rollno = rn
        self.name = n
        self.cLan = c
        self.python = p
        self.dataScience = ds

    def displayStudentReport(self):
        tot = self.cLan + self.python + self.dataScience
        per = (tot / 300) * 100

        if per >= 80:
            grade = "A"
        elif per >= 60:
            grade = "B"
        elif per >= 40:
            grade = "C"
        else:
            grade = "D"

        print("\n------------ STUDENT REPORT CARD ------------")
        print("---------------------------------------------")
        print(f"Course      : {Student.Courses}")
        print(f"Name        : {self.name}")
        print(f"Roll No.    : {self.rollno}")
        print("---------------------------------------------")
        print(f"C Language  : {self.cLan}")
        print(f"Python      : {self.python}")
        print(f"DataScience : {self.dataScience}")
        print("---------------------------------------------")
        print(f"Total Marks : {tot}/300")
        print(f"Percentage  : {per:.2f}%")
        print(f"Grade       : {grade}")
        print("---------------------------------------------")


# Creating object
s1 = Student(101, "Khushi Sharma", 88, 85, 95)

# Calling method
s1.displayStudentReport()

##INHERITENCE
class seniorStudent(student):
    def __init__(self, rn, n,c, p, ds):
        pass
    def display(self):
        print("this is senior student class")

sl = Student("KS", )
```

```python
## using default parameter
def add(n1, n2, n3=None):
    if n3 is None:
        s = n1 + n2
        print(f"Sum of {n1} and {n2} is {s}")
    else:
        s = n1 + n2 + n3
        print(f"Sum of {n1}, {n2} and {n3} is {s}")

add(5, 10)
add(2, 8, 20)
```
