```python
import numpy as np
from scipy import stats


marks = np.array([45, 67, 89, 90, 56, 78, 88, 92, 61, 74,
                  83, 95, 52, 68, 71, 85, 77, 64, 58, 81])
```

```python
# 1. Mean
print("1. Mean:", np.mean(marks))
```

```python
# 2. Median
print("Median:", np.median(marks))
```

```python
# 3. Mode
values, counts = np.unique(marks, return_counts=True)
max_count = np.max(counts)

if max_count == 1:
    print(" Mode: No mode found")
else:
    mode = values[counts == max_count]
    print("Mode:", mode)
```

```python
# 4. Maximum and Minimum
print(" Maximum:", np.max(marks))
print("  Minimum:", np.min(marks))
```

```python
# 5. Range
print("5. Range:", np.max(marks) - np.min(marks))
```

```python
# 6. Variance
print("6. Variance:", np.var(marks))
```

```python

# 7. Standard Deviation
print("7. Standard Deviation:", np.std(marks))
```

```python
# 8. First Quartile 
q1 = np.percentile(marks, 25)
print("8. Q1 (25th Percentile):", q1)
```

```python
# 9. Second Quartile (Q2)
q2 = np.percentile(marks, 50)
print("9. Q2 (50th Percentile):", q2)
```

```python
# 10. Third Quartile (Q3)
q3 = np.percentile(marks, 75)
print("10. Q3 (75th Percentile):", q3)
```

```python
# 11. Interquartile Range (IQR)
iqr = q3 - q1
print("11. IQR:", iqr)
```

```python
# 12. Students Above Mean
mean = np.mean(marks)
print("12. Students Above Mean:", np.sum(marks > mean))
```

```python
# 13. Students Below 60
print("13. Students Below 60:", np.sum(marks < 60))
```

```python
# 14. Sum and Manual Average
total = np.sum(marks)
average = total / len(marks)

print("14. Sum of Marks:", total)
print("    Manual Average:", average)
```

```python
# 15. Sorting
ascending = np.sort(marks)
descending = ascending[::-1]

print("15. Ascending Order:")
print(ascending)

print("    Descending Order:")
print(descending)

print("    Top 5 Highest Scores:")
print(descending[:5])
```
