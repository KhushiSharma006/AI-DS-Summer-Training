```python
arr = np.array([1,2,3,2,0], dtype= bool)

if np.array_equal(arr , arr[::-1]):
    print("Palindrome")
else:
    print("not palindrome")

arr
```

```python

li = []

for i in range(10):
    salary = eval(input(f"enter the salary of {i+1}th employee:"))
    li.append(salary)

arr = np.array(li)
arr

print(f"Total salary expenses: {sum(arr)} ")
print(f"Highest salary: {max(arr)} ")
print(f"Lowest salary: {min(arr)} ")


ele = eval(input(f"Search for a specific salary"))
cnt1 =0
cnt2 =0
flag=False
for i in arr:
    if i > 50000:
        cnt1 = cnt1+1
    if i< 30000:
        cnt2 = cnt2+1
    if(i == ele):   
        flag=True
    
if(flag):
    print(f"{ele} salary found")
        
else:
    print(f"salary not found")  
    
         
        



print(f"Count of employee earning more than 50000: {cnt1}")
print(f"Count of employee earning less than 30000: {cnt2}")

print(f"Salaries in ascending order:{np.sort(arr)}")
print(f"Salaries in descending order:{np.sort(arr)[:: -1]}")

print(f"original salary{arr}")
salary = np.copy(arr)
salary = salary*1.1
print(f"Updated salary:{salary}")

# Replace a salary
old_salary = float(input("\nEnter the salary to replace: "))
new_salary = float(input("Enter the new salary: "))

if old_salary in salary:
    index = np.where(salary == old_salary)[0][0]
    salary[index] = new_salary
    print("Salary updated successfully.")
else:
    print("Salary not found.")

# Display updated salary list
print("Final salary list:")
print(salary)
```
