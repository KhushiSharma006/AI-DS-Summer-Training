```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120,
                   121,122,123,124,125],

    "Name": ["Amit","Priya","Rahul","Sneha","Karan",
             "Neha","Rohit","Pooja","Vikas","Anjali",
             "Arjun","Meera","Sanjay","Riya","Deepak",
             "Komal","Nitin","Simran","Manish","Kavita",
             "Yash","Isha","Varun","Aarti","Mohit"],

    "Age": [20,21,22,20,23,
            21,22,20,24,23,
            21,22,20,23,24,
            21,22,20,23,24,
            21,22,23,20,21],

    "Course": ["B.Tech","BCA","MCA","B.Sc","B.Tech",
               "MBA","BCA","MCA","B.Tech","MBA",
               "B.Sc","BCA","MCA","MBA","B.Tech",
               "BCA","B.Sc","MBA","MCA","B.Tech",
               "BCA","MBA","B.Sc","MCA","B.Tech"],

    "Marks": [85,78,92,67,88,
              74,81,90,69,95,
              72,84,91,79,87,
              76,89,83,94,68,
              80,77,86,93,82],

    "City": ["Delhi","Mumbai","Lucknow","Jaipur","Pune",
             "Delhi","Noida","Lucknow","Indore","Bhopal",
             "Kanpur","Agra","Patna","Ranchi","Surat",
             "Chennai","Hyderabad","Kolkata","Nagpur","Delhi",
             "Jaipur","Pune","Noida","Lucknow","Mumbai"]
})

# Display DataFrame
print(df)
```

```python
#Q1. Create a line chart showing Student_ID vs Marks.
plt.plot(df["Student_ID"],  df["Marks"])
```

```python
#Q2. Create a line chart of Age vs Marks using: [marker,linewidth, linestyle, color,grid,title,legend)

plt.plot(df["Age"], df["Marks"], marker ="s", linewidth = 0.9, linestyle = "--", color = "purple", label = "Marks")
plt.grid()
plt.title("Students Records")
plt.legend()
```

```python
#Q3. Create a bar chart showing the average marks of each Course.(Hint: groupby())

avgMarks = df.groupby("Course")["Marks"].mean()
plt.bar(avgMarks.index, avgMarks.values, color = "pink")
```

```python
#Q4. Display the number of students in each City using a bar chart.


# Count students in each city
city_count = df["City"].value_counts()

# Bar Chart
plt.figure(figsize=(8,5))
city_count.plot(kind="bar")

plt.title("Number of Students in Each City")
plt.xlabel("City")
plt.ylabel("Number of Students")
plt.xticks(rotation=90)


plt.show()
```

```python
#Q5. Create a horizontal bar chart showing the top 5 students by Marks. 
top5 = df.sort_values(by = "Marks", ascending = False).head(5)


plt.barh(top5["Name"], top5["Marks"], color = "skyblue")

plt.xlabel("Student Name")
plt.ylabel("marks")
plt.title("Top 5 Stuednts list")
           
```

```python
#Q6. Create a histogram showing the distribution of Marks using: (5 bins,edgecolor, alpha)

plt.figure(figsize=(8,5))
plt.hist(df["Marks"], bins=5,color = "pink", edgecolor = "red", alpha = 0.5 )
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
#plt.grid()
```

```python
#Q7. Create a histogram of Age with:(6 bins, cumulative=True, color='orange',Scatter Plot)

plt.figure(figsize = (8,4))
plt.hist(df["Age"], bins =6, cumulative =True, color = 'orange', edgecolor = 'brown')
plt.title("Cumulative Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Cumulative Frequency")

plt.grid(axis="y")
plt.show()
```

```python
#Q8. Create a scatter plot showing Age vs Marks.
plt.scatter(df["Age"],df["Marks"])

plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Age vs Marks")

plt.show()
```

```python
#Q9 Create a scatter plot where:marker size depends on Marks,marker color depends on Age
plt.figure(figsize=(7,5))
plt.scatter(df["Age"], df["Marks"], s=df["Marks"], c=df["Age"])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
# Q10. Display the percentage of students in each Course using a pie chart.
course_count = df["Course"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(course_count,
        labels=course_count.index,
        autopct="%1.1f%%")

plt.title("Percentage of Students in Each Course")
plt.show()
```

```python
# Q11. Create a pie chart showing students from each City with:explode, shadow, startangle, percentages
city_count = df["City"].value_counts()

explode = [0.1] * len(city_count)

plt.figure(figsize=(8,8))
plt.pie(city_count,
        labels=city_count.index,
        autopct="%1.1f%%",
        explode=explode,
        shadow=True,
        startangle=90)

plt.title("Students from Each City")
plt.show()
```

```python
# Q12. Create a box plot for Marks.
plt.figure(figsize=(6,5))

plt.boxplot(df["Marks"])

plt.title("Marks Box Plot")
plt.ylabel("Marks")

plt.show()
```

```python
# Q13. Compare Marks of each Course using multiple box plots.
courses = ["B.Tech", "BCA", "MCA", "B.Sc", "MBA"]

data = [df[df["Course"] == course]["Marks"] for course in courses]

plt.figure(figsize=(8,5))

plt.boxplot(data, tick_labels=courses)

plt.title("Comparison of Marks by Course")
plt.xlabel("Course")
plt.ylabel("Marks")

plt.show()

```

```python
#q14. Sort the DataFrame by Student_ID and create an area chart of Marks.
sorted_df = df.sort_values("Student_ID")

plt.figure(figsize=(8,5))

plt.fill_between(sorted_df["Student_ID"],sorted_df["Marks"],alpha=0.6)

plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.title("Area Chart of Marks")

plt.show()
```

```python
# Q15. Create a stack plot showing cumulative marks of students grouped by three Courses (choose any three courses).
courses = ["B.Tech", "BCA", "MCA"]

data = [df[df["Course"] == course]["Marks"].reset_index(drop=True)
        for course in courses]

min_length = min(len(course_data) for course_data in data)

data_truncated = [course_data[:min_length] for course_data in data]

x = range(1, min_length + 1)

plt.figure(figsize=(10,5))

plt.stackplot(x,
              data_truncated[0],  
              data_truncated[1],  
              data_truncated[2], 
              labels=courses)

plt.title("Stack Plot of Marks by Course")
plt.xlabel("Student Number")
plt.ylabel("Marks")
plt.legend()

plt.show()
plt.show()
```
